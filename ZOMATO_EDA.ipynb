## ``` Ratings of Zomato app across the world. ```



````python
# Ratings of Zomato app across the world.
ratings = final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'}
                                                                                                            )
# Creating a Dataframe
data = pd.DataFrame(ratings)
data
````

**Observations:**
- when the ratings is between 4.5 - 4.9 ---> Excellent
- when ratings is between 4.0 - 3.4 ---> Very Good
- When ratings is between 3.5 - 3.9 ---> good
- When ratings is btween 3.0 - 3.9  ---> Average
- when rating is between 2.5 - 2.9  ---> Average
- When raing is between 2.0 to 2.4  ---> Poor

````python
# Bar Chart: Aggregate Rating vs Rating Count ------------- Not Color coded ---> for self explaination and better understanding
# plt.figure(figsize=(12, 6))
# plt.bar(data['Aggregate rating'], data['Rating Count'], color='skyblue')
# plt.xlabel('Aggregate Rating', fontsize=12)
# plt.ylabel('Rating Count', fontsize=12)
# plt.title('Aggregate Rating vs Rating Count', fontsize=14)
# plt.grid(axis='y', linestyle='--', alpha=0.7)
# plt.show()

# Color-coded Bar Chart: Aggregate Rating vs Matching Rating Color
plt.figure(figsize=(12, 6))
colors = ['green', 'blue', 'lightblue', 'orange', 'orange', 'red']
plt.bar(data['Aggregate rating'], data['Rating Count'], color=colors)
plt.xlabel('Aggregate Rating', fontsize=12)
plt.ylabel('Rating Count', fontsize=12)
plt.title('Color-coded Aggregate Rating vs Rating Count', fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
````

````python
# Based on the Observations mentioned above we create a dictionary.
categories = {
    "Excellent": (4.5, 4.9),
    "Very Good": (4.0, 4.4),
    "Good": (3.5, 3.9),
    "Average": (3.0, 3.4),
    "Average": (2.5, 2.9),
    "Poor": (2.0, 2.4)}

# Aggregate data into categories
category_counts = {category: 0 for category in categories}
for _, row in data.iterrows():
    for category, (lower, upper) in categories.items():
        if lower <= row["Aggregate rating"] <= upper:
            category_counts[category] += row["Rating Count"]
            break

# Prepare data for plotting
labels = list(category_counts.keys())
counts = list(category_counts.values())
colors = ["darkgreen", "green", "yellow", "orange", "orange", "red"]

# Plot the data
plt.figure(figsize=(10, 6))
plt.bar(labels, counts, color=colors)

# Adding labels and title
plt.xlabel("Rating Categories", fontsize=12)
plt.ylabel("Total Rating Count", fontsize=12)
plt.title("Rating Categories and Their Total Counts", fontsize=14)

# Adding labels above bars
for i, count in enumerate(counts):
    plt.text(i, count + 50, str(count), ha='center', fontsize=10)

# Display the plot
plt.ylim(0, max(counts) + 200)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
````

## `Frequency of the Ratings in colours`

````python
#Frequency of the Ratings in colours
a = sns.countplot(x = 'Rating color',data = data, palette=['blue','red','orange','yellow','green','green'])
plt.ylabel('Frequency')
plt.title('Frequency of the Ratings in colours')

# Adding frequency labels on top of the bar chart
for bar in plt.gca().patches:
    plt.text(
        bar.get_x() + bar.get_width() / 2, # X position
        bar.get_height() + 0.5,            # Y position
        int(bar.get_height()),             # Text (frequency)
        ha='center')                       # Center alignment

#     ALTERNATIVE METHOD: TO ADD BAR LABELS
# Add bar labels using bar_label
a.bar_label(a.containers[0], label_type='center')  # Access the first BarContainer for labels

plt.xticks()
plt.show()
````

**Self Note: explaination of the code**

- `plt.gca():` Stands for "get current axes," which gets the current plot where the bars are drawn.
- `.patches`: Contains all the rectangular shapes (bars) in the plot.
- This loops through each bar in the bar chart.
- `bar.get_x():` Gets the x-coordinate (left edge) of the current bar.
- `get_width():` Gets the width of the current bar.
- Adding these and dividing by 2, places the text in the horizontal center of the bar.
- `bar.get_height()`: Gets the height of the current bar, which is the frequency value.

````python
import plotly.express as px

# Count the number of restaurants by country
restaurants_by_country = final_df["Country"].value_counts().reset_index()
restaurants_by_country.columns = ["Country", "Number of Restaurants"]

# Create a map
fig = px.choropleth(
    restaurants_by_country,
    locations="Country",
    locationmode="country names",
    color="Number of Restaurants",
    title="Number of Restaurants by Country",
    color_continuous_scale="Viridis"
)

fig.update_geos(showcountries=True, showcoastlines=True, showland=True, fitbounds="locations")
fig.show()
````
